```python
import matplotlib.pyplot as plt
```

```python
def plot_prediction_distribution(predictions):
    plt.hist(predictions, bins=30, color='skyblue', edgecolor='black')

    plt.title('Prediction Confidence Distribution')
    plt.xlabel('Prediction (0=Control, 1=Dementia)')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```

```python
def plot_model_confidence(predictions):
    plt.hist(predictions, bins=20, color='skyblue', edgecolor='black')

    plt.title("Model Confidence on Unlabeled Data")
    plt.xlabel("Prediction Probability")
    plt.ylabel("Frequency")
    plt.grid(True)
    plt.show()
```
